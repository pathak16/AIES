```python
from sklearn.cluster import KMeans # Import the KMeans class
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import pandas as pd

def optimize_k_means(data,max_k):
  means=[]
  inertias=[]

  for k in range(1,max_k):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(data)
    means.append(k)
    inertias.append(kmeans.inertia_)

  #generate the elbow plot

  fig=plt.subplots(figsize=(10,5))
  plt.plot(means,inertias,'-o')
  plt.xlabel('Number of Clusters(k)')
  plt.ylabel('Inertia')
  plt.title('Elbow Method')
  plt.grid(True)
  plt.show()
```

```python
optimize_k_means(data[['Cost of Cultivation (`/Hectare) A2+FL_T', 'Cost of Cultivation (`/Hectare) C2_T','Cost of Production (`/Quintal) C2_T']],10)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data[['Cost of Cultivation (`/Hectare) A2+FL_T', 'Cost of Cultivation (`/Hectare) C2_T','Cost of Production (`/Quintal) C2_T']])
```
